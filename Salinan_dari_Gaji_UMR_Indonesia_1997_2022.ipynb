# Memilih Data

Mengecek data pada setiap kolom dan tabel,

1. Pada setiap tahun nya dari tahun 1997 sampai dengan 2022 Indonesia mengalami pemekaran provinsi, yang mana mulai yahun 1997 sampai dengan 2015 memiliki jumlah UMR di setiap tahunya berbeda maka dari itu kita akan menggunakan data yang memiliki jumlah UMR yang sama mulai tahun 2015 sampai dengan 2022

2. Mari kita buat datset ringkasan gaji dari 1997 sampai dengan 2022 Indonesia, jika kita lihat perhitungan rata-rata gaji tidak akurat karena di beberapa tahun UMR Indonesia berbeda-beda jumlahnya

```r
#Mengecek jumlah UMR pada setiap tahun berdasarkan kota berapa banyak UMR pada setiap tahun nya
summary_many_UMR_in_year_by_Region <- salary_id %>%
  filter(YEAR >= "1997") %>%
  arrange(YEAR) %>%
  group_by(YEAR) %>%
  summarize(REGION=n())
```

```r
#Ringkasan UMR berdasarkan Provinsi Tahun 1997-2022
summary_UMR_by_Region_1997_s.d_2022 <- salary_id %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY)) %>%
  arrange(REGION)
```

# Membuat Ringkasan
Ringkasan perbandingan gaji di setiap Gugusan Pulau Indonesia sejak 2017 s.d 2022

1. Pulau Jawa yaitu Provinsi DKI Jakarta, Banten, Jawa Barat, Jawa Tengah, Jawa Timur, DI Yogyakarta,

2. Pulau Sumatera yaitu Provinsi Aceh, Sumatera Utara, Sumatera Barat, Riau, Kep. Riau, Jambi, Bengkulu, Sumatera Selatan, Kep. Bangka Belitung, Lampung

3. Pulau Kalimantan yaitu Provinsi  Kalimantan Barat, Kalimantan Utara, Kalimantan Timur,Kalimantan Selatan,Kalimantan Tengah

4. Pulau Sulawesi yaitu Provinsi Sulawesi Barat, Sulawesi Utara, Sulawesi Tenggara, Sulawesi Selatan, Sulawesi Tengah, Gorontalo

5. Kepulauan Sunda Kecil yaitu Provinsi Bali, Nusa Tenggara Barat, Nusa Tenggara Timur

6. Kepulauan Maluku yaitu Provinsi Maluku, Maluku Utara

7. Pulau Papua yaitu Provinsi Papua, Papua Barat

```r
#Ringkasan UMR Provinsi Pulau Jawa 2017 - 2022
Summary_UMR_by_Region_Java_2017_s.d_2022 <- filter(salary_id, REGION %in% c("DKI JAKARTA", "BANTEN", "JAWA BARAT",
                                                              "JAWA TENGAH", "DI YOGYAKARTA", "JAWA TIMUR"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Pulau Sumatera 2017 - 2022
Summary_UMR_by_Region_Sumatera_2017_s.d_2022 <- filter(salary_id, REGION %in% c("ACEH", "SUMATERA UTARA", "SUMATERA BARAT",
                                                              "RIAU", "KEP. RIAU", "JAMBI", "BENGKULU", "SUMATERA SELATAN",
                                                              "KEP. BANGKA BELITUNG", "LAMPUNG"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Pulau Kalimantan 2017 - 2022
Summary_UMR_by_Region_Kalimantan_2017_s.d_2022 <- filter(salary_id, REGION %in% c("KALIMANTAN BARAT", "KALIMANTAN UTARA", "KALIMANTAN TIMUR",
                                                                                  "KALIMANTAN SELATAN","KALIMANTAN TENGAH"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Pulau Sulawesi 2017 - 2022
Summary_UMR_by_Region_Sulawesi_2017_s.d_2022 <- filter(salary_id, REGION %in% c("SULAWESI BARAT", "SULAWESI UTARA", "SULAWESI TENGGARA",
                                                                                  "SULAWESI SELATAN","SULAWESI TENGAH", "GORONTALO"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Sunda Kecil 2017 - 2022
Summary_UMR_by_Region_Sunda_Kecil_2017_s.d_2022 <- filter(salary_id, REGION %in% c("BALI", "NUSA TENGGARA BARAT", "NUSA TENGGARA TIMUR"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Maluku 2017 - 2022
Summary_UMR_by_Region_Maluku_2017_s.d_2022 <- filter(salary_id, REGION %in% c("MALUKU", "MALUKU UTARA"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```

```r
#Ringkasan UMR Provinsi Papua 2017 - 2022
Summary_UMR_by_Region_Papua_2017_s.d_2022 <- filter(salary_id, REGION %in% c("PAPUA", "PAPUA BARAT"), YEAR >= "2017") %>%
  group_by(REGION) %>%
  summarize(Rata_rata_Gaji = mean(SALARY),
            Gaji_Terendah = min(SALARY),
            Gaji_Tertinggi = max(SALARY),
            Jumlah_Besaran_Gaji_Tiap_Tahun = n()) %>%
  arrange(REGION)
```
